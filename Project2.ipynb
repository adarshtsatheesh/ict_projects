8. Course Registration System
Design a University Course Registration System.
The system should allow creation of courses with credits and seat limits, student
registration, and course enrollment.
Constraints include maximum credit limits per student, prevention of duplicate
registrations, and seat availability enforcement. Summary reports for students and courses
must be generated.

```python
#data
courses = {}     # course_name : {credits, seats, enrolled}
students = {}    # student_name : {courses, total_credits}

MAX_CREDITS = 6   # maximum credits per student

#add course
def add_course():
    cname = input("Enter course name: ")
    credits = int(input("Enter credits: "))
    seats = int(input("Enter seat limit: "))

    courses[cname] = {
        "credits": credits,
        "seats": seats,
        "enrolled": 0
    }
    print("Course added successfully")

# ---------------- REGISTER STUDENT ----------------
def register_student():
    sname = input("Enter student name: ")

    if sname in students:
        print("Student already registered")
    else:
        students[sname] = {
            "courses": [],
            "total_credits": 0
        }
        print("Student registered successfully")

# ---------------- ENROLL COURSE ----------------
def enroll_course():
    sname = input("Enter student name: ")
    cname = input("Enter course name: ")

    if sname not in students:
        print("Student not registered")
        return

    if cname not in courses:
        print("Course not found")
        return

    # duplicate course check
    if cname in students[sname]["courses"]:
        print("Student already enrolled in this course")
        return

    # credit limit check
    course_credits = courses[cname]["credits"]
    if students[sname]["total_credits"] + course_credits > MAX_CREDITS:
        print("Credit limit exceeded")
        return

    # seat availability check
    if courses[cname]["enrolled"] >= courses[cname]["seats"]:
        print("No seats available")
        return

    # enroll
    students[sname]["courses"].append(cname)
    students[sname]["total_credits"] += course_credits
    courses[cname]["enrolled"] += 1

    print("Enrollment successful")

# ---------------- STUDENT SUMMARY ----------------
def student_summary():
    print("\n--- Student Summary ---")
    for s in students:
        print("Name:", s)
        print("Courses:", students[s]["courses"])
        print("Total Credits:", students[s]["total_credits"])
        print()

# ---------------- COURSE SUMMARY ----------------
def course_summary():
    print("\n--- Course Summary ---")
    for c in courses:
        print("Course:", c)
        print("Credits:", courses[c]["credits"])
        print("Seats:", courses[c]["seats"])
        print("Enrolled:", courses[c]["enrolled"])
        print()

# ---------------- MENU ----------------
while True:
    print("\n--- COURSE REGISTRATION SYSTEM ---")
    print("1. Add course")
    print("2. Register student")
    print("3. Enroll student in course")
    print("4. Student summary report")
    print("5. Course summary report")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_course()
    elif choice == "2":
        register_student()
    elif choice == "3":
        enroll_course()
    elif choice == "4":
        student_summary()
    elif choice == "5":
        course_summary()
    elif choice == "6":
        print("Program Ended")
        break
    else:
        print("Invalid choice")
```
