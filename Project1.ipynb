10. Log File Analyzer
Develop a Log File Analyzer for simulated system logs.
The program should parse text-based log entries, categorize events, count error types,
analyze user activity, and filter logs based on time or severity.
String splitting, pattern matching, and dictionary-based aggregation must be used. The
application should present analytical summaries via a menu-driven interface

```python
# Create a simulated log file
file = open("logs.txt", "w")

file.write("10:15 | ERROR | Aadhil | Failed login\n")
file.write("10:16 | INFO | Niranj | File uploaded\n")
file.write("10:17 | WARNING | Irfana | Disk full\n")

file.close()
```

```python
# load logs
def load_logs():
    logs = []
    file = open("logs.txt", "r")
    for line in file:
        logs.append(line.strip())
    file.close()
    return logs

logs = load_logs()

# Add new entry
def add_log():
    time = input("Enter time (HH:MM): ")
    severity = input("Enter severity (INFO/WARNING/ERROR): ")
    user = input("Enter username: ")
    message = input("Enter message: ")

    log = time + " | " + severity + " | " + user + " | " + message
    logs.append(log)

    file = open("logs.txt", "a")
    file.write(log + "\n")
    file.close()

    print("Log added successfully")

#show all logs
def show_logs():
    print("\n----- Logs -----")
    for log in logs:
        print(log)

#categorize events
def categorize_events():
    categories = {}

    for log in logs:
        parts = log.split(" | ")
        severity = parts[1]

        if severity in categories:
            categories[severity] += 1
        else:
            categories[severity] = 1

    print("\nEvent Categories:")
    for k in categories:
        print(k, ":", categories[k])

#count error type
def count_error_types():
    errors = {}

    for log in logs:
        parts = log.split(" | ")
        if parts[1] == "ERROR":
            message = parts[3]
            if message in errors:
                errors[message] += 1
            else:
                errors[message] = 1

    print("\nError Types:")
    for k in errors:
        print(k, ":", errors[k])

# analyze user activity
def analyze_user_activity():
    users = {}

    for log in logs:
        parts = log.split(" | ")
        user = parts[2]

        if user in users:
            users[user] += 1
        else:
            users[user] = 1

    print("\nUser Activity:")
    for k in users:
        print(k, ":", users[k])

#filter by serverity function
def filter_by_severity():
    level = input("Enter severity (INFO/WARNING/ERROR): ")

    print("\nFiltered Logs:")
    for log in logs:
        parts = log.split(" | ")
        if parts[1] == level:
            print(log)

#filter by time function
def filter_by_time():
    time_input = input("Enter time (HH:MM): ")

    print("\nFiltered Logs:")
    for log in logs:
        parts = log.split(" | ")
        if parts[0] == time_input:
            print(log)

#menu
while True:
    print("\n----- BASIC LOG ANALYZER -----")
    print("1. Add new log entry")
    print("2. Show all logs")
    print("3. Categorize events")
    print("4. Count error types")
    print("5. Analyze user activity")
    print("6. Filter logs by severity")
    print("7. Filter logs by time")
    print("8. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_log()
    elif choice == "2":
        show_logs()
    elif choice == "3":
        categorize_events()
    elif choice == "4":
        count_error_types()
    elif choice == "5":
        analyze_user_activity()
    elif choice == "6":
        filter_by_severity()
    elif choice == "7":
        filter_by_time()
    elif choice == "8":
        print("Program Ended")
        break
    else:
        print("Invalid choice")
```
